```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

path = r"C:\Users\jcp\Documents\PYTHON LABO\Regres_linear\povo_2.csv"
df = pd.read_csv(path,encoding='ISO-8859-1')
# print dataframe
print("============================= priting data frame" )
print(df.head())
#descrbe a dataframe
print("============================== describing data frame")
print(df.describe())
#select onl feature "idade" e "anos"
idades = df[['IDADE']]
ano = df[['ANOS']]
print("============================== selecting only feature 'idade' e 'anos'")
print(idades.head())
print(ano.head())

"""#Dividir o dataset em dados de treinamento e dados de teste
neste casos vamos usar o train_test_split do scikitlearn
"""
idades_treino, idades_test, ano_treino, ano_test = train_test_split(idades, ano, test_size=0.2, random_state=42)
print("============================= beaking data frame into 2 part of 20% and 80%")
print(type(idades_treino))
"""#Exibir a correlação entre as features do dataset de treinamento"""
print("============================= correlação entre as features do dataset de treinamento")
plt.scatter(idades_treino, ano_treino, color = 'blue')
plt.xlabel('Idade')
plt.ylabel('Ano')
plt.show()

"""# Vamos treinar o modelo de regressão linear"""

# CRIAR UM MODELO DE TIPO DE REGRESSÃO LINEAR
modelo =  linear_model.LinearRegression()

# TREINAR O MODELO USANDO O DATASET DE TESTE(TREINO)
# PARA ENCONTRAR O VALOR DE A E B (Y = A + B.X)
modelo.fit(idades_treino,ano_treino)

"""#Exibir os coeficientes (A e B)"""
print("============================== Exibir os coeficientes (A e B)")
print("A-> intercepto :", modelo.intercept_)
print("B-> inclinação :", modelo.coef_)

"""# Vamos exibir a nossa reta de regressão no dataset de treino"""
print("============================== exibir a nossa reta de regressão no dataset de treino")
plt.scatter(idades_treino, ano_treino, color = 'blue')
plt.plot(idades_treino, modelo.coef_[0][0]*idades_treino + modelo.intercept_[0], '-r')
plt.xlabel('Idade')
plt.ylabel('Ano')
plt.show()


#Primeiro a gente tem que fazer as predições usando o modelo e base de teste
predicoesAnos = modelo.predict(idades_test)

"""# Vamos exibir a nossa reta de regressão no dataset de teste"""
print("============================== exibir a nossa reta de regressão no dataset de test")
plt.scatter(idades_treino, ano_treino, color = 'blue')
plt.plot(idades_test, modelo.coef_[0][0]*idades_test + modelo.intercept_[0], '-r')
plt.xlabel('Idade')
plt.ylabel('Ano')
plt.show()

"""# Vamos avaliar o modelo"""

#Agora é mostrar as métricas
print("Soma dos Erros ao Quadrado (SSE): %.2f " % np.sum((predicoesAnos - ano_test)**2))
print("Erro Quadrático Médio (MSE): %.2f" % mean_squared_error(ano_test, predicoesAnos))
print("Erro Médio Absoluto (MAE): %.2f" % mean_absolute_error(ano_test, predicoesAnos))
print ("Raiz do Erro Quadrático Médio (RMSE): %.2f " % sqrt(mean_squared_error(ano_test, predicoesAnos)))
print("R2-score: %.2f" % r2_score(ano_test, predicoesAnos) )
```
