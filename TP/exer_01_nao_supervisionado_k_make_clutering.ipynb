```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#x = np.array([[1,2],[3,4],[5,6]])
x,y = make_blobs(n_samples=100, centers=3, cluster_std=1.0, random_state=42)
print("*"*20)
#***** plottar os 3 clusters brutos sem agrupamento dos pontos semelhentes******
plt.scatter(x[:,0], x[:,1])
plt.show()
print("*"*20)

model = KMeans(n_clusters=3, n_init=10, max_iter=200, init='k-means++')
model.fit(x)
labels = model.labels_
print(labels)
model.predict(x)
#***** plottar os 3 clusters ******
plt.scatter(x[:,0], x[:,1], c=model.predict(x))

#exibir les centroid
centroids = model.cluster_centers_
print("les centroids et les clusters:", centroids)

#plotter les centroids
#********plottar os 3 centroids dos 3 clusters **********"*
print("******mostrar os 3 centroids dos clusters*******")
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c='r')
print("*"*20)

#calculer 'inertia': la somme de distance la plus courte entre un point d'un cluster et un centroid
inertia = model.inertia_
print("inertia : ", inertia)
#ou encore:
score = model.score(x)
print("inertia ou score : ", score)
```
